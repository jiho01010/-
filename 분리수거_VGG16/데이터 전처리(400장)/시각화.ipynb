```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# 
save_dir = "plots"
os.makedirs(save_dir, exist_ok=True)

#  
df = pd.read_csv('metrics.csv')
x = df['epoch'] if 'epoch' in df.columns else df.index
x_label = 'Epoch' if 'epoch' in df.columns else 'Index'

#  
compare_pairs = [
    ('train_loss', 'val_loss'),
    ('train_acc',  'val_acc'),
]

#  + 
for train_col, val_col in compare_pairs:
    if train_col in df.columns and val_col in df.columns:
        plt.figure(figsize=(8, 4))
        plt.plot(x, df[train_col], marker='o', label='Test')
        plt.plot(x, df[val_col], marker='s', label='Validation')

        metric = train_col.split('_', 1)[1].capitalize()  # 'loss'  'acc'
        plt.xlabel(x_label)
        plt.ylabel(metric)
        plt.title(f'Test vs Validation {metric}')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()

        out_path = os.path.join(save_dir, f"{metric.lower()}.png")
        plt.savefig(out_path, dpi=300)
        print(f"[Saved] {out_path}")

        plt.show()
    else:
        print(f"[Skip] Missing columns: {train_col}, {val_col}")
```
